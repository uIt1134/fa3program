```python
score = int(input("Enter student's exam score: "))

if score > 100:
    print("Max score is 100.")

elif score > 74:
    print("Passed")
else:
    print("Failed")
```

```python
student_name = input("Enter student's name: ")
score = int(input("Enter student's score: "))

if 90 <= score <= 100:
    grade = "A"
elif 80 <= score <= 89:
    grade = "B"
elif 70 <= score <= 79:
    grade = "C"
else:
    grade = "F"

print(f"Student: {student_name}")
print(f"Score: {score}")
print(f"Grade: {grade}")
```
